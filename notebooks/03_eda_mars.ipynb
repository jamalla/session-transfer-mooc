Goal:
- Load the MARS ratings/interaction file.
- Inspect columns (user_id, item_id, timestamps, implicit/explicit signals).
- Plan how to convert to (user_id, session_id, item_id, timestamp, interaction_type).

```python
import pandas as pd

path = "../data/raw/mars/explicit_ratings_en.csv"  # or implicit/explicit file
df = pd.read_csv(path)

df.head()
```

```python
print(df.columns)
print(df.dtypes)
print("Users:", df["user_id"].nunique())
print("Items:", df["item_id"].nunique())
print("Interactions:", len(df))
```

```python
df["created_at"] = pd.to_datetime(df["created_at"])
df.sort_values(["user_id", "created_at"], inplace=True)
df.head()
```

```python
df_uni = pd.DataFrame({
    "dataset": "mars",
    "user_id": df["user_id"].astype(str),
    "session_id": df["user_id"].astype(str),  # temp; will split by time gaps later
    "item_id": df["item_id"].astype(str),
    "timestamp": df["created_at"],
    "interaction_type": "explicit_rating"  # we can refine: rating or watch percentage
})
```

```python
df_uni.to_parquet("../data/processed/mars_interactions.parquet", index=False)
```
